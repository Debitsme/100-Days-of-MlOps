``Black and Ruff`` are essential Python code quality tools that work beautifully together. Black is an automatic code formatter that enforces a strict, consistent style across your codebase—removing manual formatting decisions and ensuring every file looks the same. Ruff, on the other hand, is a blazing-fast linter (written in Rust) that scans your code for bugs, style violations, and unused imports, with the ability to auto-fix many issues instantly. While Black handles formatting to make your code pretty, Ruff catches errors and enforces best practices—together, they keep your Python code clean, consistent, and error-free with minimal effort.



``ruff check src/
``

``black --check src/``

.toml file

[project]
name = "fraud-detection"
version = "0.1.0"

[tool.ruff]
line-length = 120
 
[tool.ruff.lint]
select = ["E", "F", "W", "I"]

[tool.black]
line-length = 120
