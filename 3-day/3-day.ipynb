# What is UV in python
uv is a Python package manager and environment tool. It installs and manages dependencies, similar to pip, but it is designed to be faster and more consistent
# What is a lock file
A lockfile is a file that records the exact versions of every package being used, including indirect dependencies.
``pandas==2.2.1``
``numpy==1.26.4``
