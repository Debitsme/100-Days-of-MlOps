A Makefile is a file that stores commands (called targets) so you can run them with the make command instead of typing the full commands manually each time.

```python
# fraud-detection Makefile

.PHONY: setup data train test clean all

setup:
	python3 -m venv mlops-venv && mlops-venv/bin/pip install -r requirements.txt

data:
	python src/data/process_data.py

train:
	python src/models/train.py

test:
	pytest tests/

clean:
	find . -type d -name __pycache__ -exec rm -rf {} +
	rm -rf .pytest_cache
	rm -rf models/*

all: setup data train test
```

`` # make all``

#Task
All six target names must be declared as .PHONY so that Make never confuses them with files of the same name.
### Solution
.PHONY: setup data train test clean all

By marking them as .PHONY, you're telling make:

"These are tasks. Always run them when requested."
